```python
# Import Libraries
import pandas as pd
import numpy as np

from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Import Dataset
books=pd.read_csv(r"C:\Users\safreena\Downloads\book (2).csv",encoding='Latin1')
books
```

```python
# Display summary statistics for the numerical columns
print(books.describe())
```

## EDA

```python
# Create a count plot of book ratings
sns.countplot(x='Book.Rating', data=books)

# Add labels and title
plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Distribution of Book Ratings')
plt.show()
```

```python
# Group the data by book title and calculate the total number of ratings for each book
book_ratings = books.groupby('Book.Title')['Book.Rating'].count().reset_index(name='Total Ratings')

# Sort the data in descending order by total ratings and select the top 10 books
top_10_books = book_ratings.sort_values(by='Total Ratings', ascending=False).head(10)

# Create a bar plot of the top 10 most rated books
sns.barplot(x='Total Ratings', y='Book.Title', data=top_10_books)

# Add labels and title
plt.xlabel('Total Ratings')
plt.ylabel('Book Title')
plt.title('Top 10 Most Rated Books')
plt.show()
```

```python
books2=books.iloc[:,1:]
books2
```

```python
# Sort by User IDs
books2.sort_values(['User.ID'])
```

```python
# number of unique users in the dataset
len(books2['User.ID'].unique())
```

```python
# number of unique books in the dataset
len(books2['Book.Title'].unique())
```

```python
# converting long data into wide data using pivot table
books3=books2.pivot_table(index='User.ID',columns='Book.Title',values='Book.Rating').reset_index(drop=True)
books3
```

```python
# Replacing the index values by unique user Ids
books3.index=books2['User.ID'].unique()
books3
```

```python
# Impute those NaNs with 0 values
books3.fillna(0,inplace=True)
books3
```

```python
# Calculating Cosine Similarity between Users on array data
user_sim=1-pairwise_distances(books3.values,metric='cosine')
user_sim
```

```python
# Store the results in a dataframe format
user_sim2=pd.DataFrame(user_sim)
user_sim2
```

```python
# Set the index and column names to user ids 
user_sim2.index=books2['User.ID'].unique()
user_sim2.columns=books2['User.ID'].unique()
user_sim2
```

```python
# Nullifying diagonal values
np.fill_diagonal(user_sim,0)
user_sim2
```

```python
# Most Similar Users
user_sim2.idxmax(axis=1)
```
